```python
article_tweets = {}
total_tweets = 0
for article in os.listdir('/content/politifact/fake/'):
    if article == ".DS_Store":
        continue
    tweets_list =  os.listdir(f'/content/politifact/fake/{article}/tweets')
    article_tweets[article] =tweets_list
    total_tweets += len(tweets_list)
```

```python
total_tweets
```

```python
attributes = ["created_at", "id", "full_text", "source", "geo", "coordinates", "place", "retweet_count", "favorite_count", "favorited", "retweeted","possibly_sensitive", "possibly_sensitive_appealable", "lang"]
user_attr = ["id", "name","screen_name","location","description","url","protected","followers_count","friends_count","listed_count","created_at","favourites_count"]
```

```python
store = []
for article, tweets in tqdm.tqdm(article_tweets.items()):
    for tweet_file in tweets:
        try:
            with open(f"/content/politifact/fake/{article}/tweets/{tweet_file}", 'r') as f:
                tweet = json.load(f)
                temp = {}
                temp['dataset'] = 'politifact'
                temp['label'] = 'fake'
                temp['article_id'] = article
                temp['user'] = {}
                for attribute in attributes:
                    try:
                        temp[attribute] = tweet[attribute]
                    except:
                        temp[attribute] = "NULL"
                
                for attribute in user_attr:
                    try:
                        temp["user"][attribute] = tweet["user"][attribute]
                    except:
                        temp["user"][attribute] = "NULL"

                store.append(temp)
        except:
            pass

with open('/content/gdrive/MyDrive/BT4222/Data/tweets/politifact_fake_neil.json', 'w') as f:
    json.dump(store,f)
```

### Collect all json into one

```python
names = ['xp','alina','jk','neil']
labels = ['real','fake']
datasets = ['gossipcop', 'politifact']
```

```python
base_path = '/content/gdrive/MyDrive/BT4222/Data/tweets'
```

```python
gossipcop = []
politifact = []

for dataset in datasets:
    for label in labels:
        for name in names:
            try:
                with open(f"{base_path}/{dataset}_{label}_{name}.json", 'r') as f:
                    tweets = json.load(f)
                    if dataset == "gossipcop":
                        gossipcop.extend(tweets)
                    else:
                        politifact.extend(tweets)
            except:
                pass
```
