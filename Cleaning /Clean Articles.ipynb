```python
import os
import json
import pandas as pd
```

```python
x = os.listdir('/content/fakenewsnet_dataset/politifact/real/')
```

```python
#store = []
#no_article =0
for y in x:
  temp = {}
  try:
    with open(f"/content/fakenewsnet_dataset/politifact/real/{y}/news content.json", 'r') as f:
      article = json.load(f)
      temp['id'] = y
      temp['text'] = article['text']
      temp['title'] = article['title']
      
      temp['publish_date'] = article['publish_date']
      temp['source'] = article['source']
      temp['url'] = article['canonical_link']
      
      try:
        temp['publisher'] = article['meta_data']['og']['site_name']
      except:
        pass

      try:
        temp['published_time'] = article['meta_data']['article']['published_time']
      except:
        pass    


      temp['dataset'] = 'politifact'
      temp['label'] = 'real'

      store.append(temp)
  except:
    no_article+=1
```

```python
len(store)
```

```python
with open('politifact_v2.json', 'w') as f:
  json.dump(store,f)
```
